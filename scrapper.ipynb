```python
# Gerekli kütüphanelerin eklenmesi
!pip install ntscraper
import pandas as pd
from ntscraper import Nitter
```

```python
# Kazıcı
scraper = Nitter()
```

```python
# Tweetleri çekmek için fonksiyon
def scrap_tweets(name, modes, quantity):
  tweets = scraper.get_tweets(name, mode = modes, number = quantity)
  final_tweets = []

  for tweet in tweets['tweets']:
    data = [tweet['text']]
    final_tweets.append(data)
    data = pd.DataFrame(final_tweets, columns=['text'])

  return data
```

```python
# Tweet çekmek için arama sorgusu
data = scrap_tweets('oppenheimer watched lang:en since:2023-07-21 until:2023-08-01', 'term', 1000)
```

```python
# Çekilen tweetlerin bir .csv dosyasına kaydedilmesi
data.to_csv("watched.csv", encoding="utf-8", index = "false")
```
